```python
match_win = df['winner'].value_counts()
sns.barplot(y = match_win.index,x =  match_win.values, palette ='magma')
plt.title("Most match win by team")
```

```python
sns.countplot(x = df['decision'], palette='inferno')
plt.title("Toss decision Trends")
```

```python
check = df[df['toss_winner'] == df['winner']]['id'].count()
percentage = (check * 100)/df.shape[0]
percentage.round(2)
```

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x = count.values, y = count.index, palette='rainbow')
plt.title("top 10 players of the match")
```

```python
high = df['most_runs'].value_counts().head(2)
high
```

```python
sns.barplot(x = high.values, y = high.index, palette='rainbow')
plt.title("top 2 players who have highest run")
```

```python
wickets = df['most_wkts'].value_counts().head(5)
wickets
```

```python
sns.barplot(x = wickets.values, y = wickets.index, palette='cool')
plt.title("top 5 players who have most wkts in one match")
```
